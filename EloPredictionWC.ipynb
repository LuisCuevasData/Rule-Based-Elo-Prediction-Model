```python
# ── STEP 4: Match Prediction Engine (FIXED) ───────────────────────

def predict_match(team_a, team_b):
    form_a = recent_form.get(team_a, {'win_rate':0.4, 'avg_scored':1.2, 'avg_conceded':1.4})
    form_b = recent_form.get(team_b, {'win_rate':0.4, 'avg_scored':1.2, 'avg_conceded':1.4})
    wc_a   = wc_history.get(team_a, default_wc)
    wc_b   = wc_history.get(team_b, default_wc)
    elo_a  = elo_ratings.get(team_a, 1400.0)
    elo_b  = elo_ratings.get(team_b, 1400.0)

    # Component 1: Elo
    elo_prob_a = expected_score(elo_a, elo_b)

    # Component 2: Recent form
    form_score_a = (form_a['win_rate'] * 0.5 +
                    form_a['avg_scored'] / 3.0 * 0.3 +
                    (1 - form_a['avg_conceded'] / 3.0) * 0.2)
    form_score_b = (form_b['win_rate'] * 0.5 +
                    form_b['avg_scored'] / 3.0 * 0.3 +
                    (1 - form_b['avg_conceded'] / 3.0) * 0.2)
    form_prob_a = form_score_a / (form_score_a + form_score_b)

    # Component 3: WC history
    wc_score_a = (wc_a['wc_win_rate'] * 0.5 +
                  wc_a['avg_wc_scored'] / 3.0 * 0.3 +
                  (1 - wc_a['avg_wc_conceded'] / 3.0) * 0.2)
    wc_score_b = (wc_b['wc_win_rate'] * 0.5 +
                  wc_b['avg_wc_scored'] / 3.0 * 0.3 +
                  (1 - wc_b['avg_wc_conceded'] / 3.0) * 0.2)
    wc_prob_a = wc_score_a / (wc_score_a + wc_score_b)

    # Composite win probability for A (before draw adjustment)
    raw_prob_a = (0.40 * elo_prob_a +
                  0.35 * form_prob_a +
                  0.25 * wc_prob_a)

    # Draw probability: peaks at 27% when evenly matched, drops off sharply
    margin = abs(raw_prob_a - 0.5)
    draw_prob = max(0.05, 0.27 - margin * 1.8)

    # Redistribute remaining probability
    remaining = 1 - draw_prob
    win_prob_a = raw_prob_a * remaining / (raw_prob_a + (1 - raw_prob_a))
    win_prob_b = (1 - raw_prob_a) * remaining / (raw_prob_a + (1 - raw_prob_a))

    # Determine outcome using CLEAR thresholds
    # A team must have meaningfully higher win prob to avoid a draw result
    if win_prob_a > win_prob_b + 0.12:
        outcome = 'A'
    elif win_prob_b > win_prob_a + 0.12:
        outcome = 'B'
    else:
        outcome = 'D'

    # Expected goals
    exp_goals_a = round((form_a['avg_scored'] * 0.5 +
                         wc_a['avg_wc_scored'] * 0.3 +
                         form_b['avg_conceded'] * 0.2), 2)
    exp_goals_b = round((form_b['avg_scored'] * 0.5 +
                         wc_b['avg_wc_scored'] * 0.3 +
                         form_a['avg_conceded'] * 0.2), 2)

    # Scale goals by outcome
    if outcome == 'A':
        exp_goals_a = round(exp_goals_a * 1.1, 2)
        exp_goals_b = round(exp_goals_b * 0.7, 2)
    elif outcome == 'B':
        exp_goals_a = round(exp_goals_a * 0.7, 2)
        exp_goals_b = round(exp_goals_b * 1.1, 2)
    else:
        avg = round((exp_goals_a + exp_goals_b) / 2, 2)
        exp_goals_a = exp_goals_b = avg

    return {
        'team_a': team_a, 'team_b': team_b,
        'win_prob_a': round(win_prob_a * 100, 1),
        'draw_prob':  round(draw_prob  * 100, 1),
        'win_prob_b': round(win_prob_b * 100, 1),
        'outcome': outcome,
        'exp_goals_a': exp_goals_a,
        'exp_goals_b': exp_goals_b,
    }

def simulate_group(group_name, teams):
    print(f"\n{'='*62}")
    print(f"  GROUP {group_name}: {' | '.join(teams)}")
    print(f"{'='*62}")

    standings = {t: {'pts':0,'w':0,'d':0,'l':0,'gf':0.0,'ga':0.0} for t in teams}

    matches = []
    for i in range(len(teams)):
        for j in range(i+1, len(teams)):
            team_a, team_b = teams[i], teams[j]
            result = predict_match(team_a, team_b)
            matches.append(result)

            ga, gb = result['exp_goals_a'], result['exp_goals_b']

            if result['outcome'] == 'A':
                standings[team_a]['pts'] += 3
                standings[team_a]['w']   += 1
                standings[team_b]['l']   += 1
            elif result['outcome'] == 'B':
                standings[team_b]['pts'] += 3
                standings[team_b]['w']   += 1
                standings[team_a]['l']   += 1
            else:
                standings[team_a]['pts'] += 1
                standings[team_b]['pts'] += 1
                standings[team_a]['d']   += 1
                standings[team_b]['d']   += 1

            standings[team_a]['gf'] += ga
            standings[team_a]['ga'] += gb
            standings[team_b]['gf'] += gb
            standings[team_b]['ga'] += ga

    for t in teams:
        standings[t]['gd'] = standings[t]['gf'] - standings[t]['ga']

    print("\n  MATCH PREDICTIONS:")
    print(f"  {'Home':<25} Win%   Draw%  Win%  {'Away'}")
    print(f"  {'-'*60}")
    for m in matches:
        print(f"  {m['team_a']:<25} {m['win_prob_a']:5.1f}% "
              f"{m['draw_prob']:5.1f}% {m['win_prob_b']:5.1f}%  {m['team_b']}")

    sorted_teams = sorted(teams,
        key=lambda t: (standings[t]['pts'],
                       standings[t]['gd'],
                       standings[t]['gf']), reverse=True)

    print(f"\n  PREDICTED STANDINGS:")
    print(f"  {'Team':<28} Pts  W  D  L   GF   GA   GD")
    print(f"  {'-'*60}")
    for i, t in enumerate(sorted_teams):
        s = standings[t]
        gd = s['gf'] - s['ga']
        if i == 0:
            prefix = "🥇"
        elif i == 1:
            prefix = "🥈"
        else:
            prefix = "  "
        print(f"  {prefix} {t:<26} {s['pts']:3}  {s['w']}  "
              f"{s['d']}  {s['l']}  {s['gf']:4.1f}  {s['ga']:4.1f}  {gd:+.1f}")

    return sorted_teams, standings

# ── STEP 5: Run all 12 groups ──────────────────────────────────────
groups = {
    'A': ['Mexico', 'South Korea', 'South Africa', 'Czechia'],
    'B': ['Canada', 'Bosnia and Herzegovina', 'Qatar', 'Switzerland'],
    'C': ['Brazil', 'Morocco', 'Haiti', 'Scotland'],
    'D': ['United States', 'Paraguay', 'Australia', 'Turkey'],
    'E': ['Spain', 'Uzbekistan', 'New Zealand', 'Jordan'],
    'F': ['Portugal', 'Algeria', 'Egypt', 'Ivory Coast'],
    'G': ['Argentina', 'Curacao', 'Panama', 'Ecuador'],
    'H': ['France', 'Senegal', 'Norway', 'Colombia'],
    'I': ['Germany', 'Netherlands', 'Saudi Arabia', 'Cabo Verde'],
    'J': ['England', 'Croatia', 'DR Congo', 'Ghana'],
    'K': ['Japan', 'Belgium', 'Austria', 'Sweden'],
    'L': ['Uruguay', 'Iran', 'South Korea', 'Tunisia'],
}

group_results = {}
all_third_place = []

for group_name, teams in groups.items():
    sorted_teams, standings = simulate_group(group_name, teams)
    group_results[group_name] = {
        'ranked': sorted_teams,
        'standings': standings
    }
    # Collect third place teams with their points/gd for best-of-8 selection
    third = sorted_teams[2]
    all_third_place.append({
        'team': third,
        'group': group_name,
        'pts': standings[third]['pts'],
        'gd': standings[third]['gd'],
        'gf': standings[third]['gf'],
    })

# ── Best 8 third-place teams advance ──────────────────────────────
best_thirds = sorted(all_third_place,
    key=lambda x: (x['pts'], x['gd'], x['gf']), reverse=True)[:8]
best_third_teams = {t['team'] for t in best_thirds}

print("\n\n" + "="*62)
print("  GROUP STAGE SUMMARY")
print("="*62)
print(f"\n  {'Grp'} {'1st':<28} {'2nd':<28} {'3rd (best 8?)'}")
print(f"  {'-'*80}")
for g in sorted(groups.keys()):
    r = group_results[g]['ranked']
    third_note = "✅" if r[2] in best_third_teams else "❌"
    print(f"   {g}  {r[0]:<28} {r[1]:<28} {r[2]} {third_note}")

print(f"\n  BEST 8 THIRD-PLACE TEAMS ADVANCING:")
for t in best_thirds:
    print(f"  ✅ {t['team']:<28} (Group {t['group']}) "
          f"Pts:{t['pts']} GD:{t['gd']:+.1f} GF:{t['gf']:.1f}")
```

```python
# ── STEP 6: Round of 32 & Knockout Stage ──────────────────────────

def knockout_match(team_a, team_b, round_name):
    """In knockouts there are no draws — extra time/penalties if needed"""
    result = predict_match(team_a, team_b)

    # In knockouts, draw goes to slight favorite
    if result['outcome'] == 'D':
        if result['win_prob_a'] >= result['win_prob_b']:
            result['outcome'] = 'A'
        else:
            result['outcome'] = 'B'

    winner = team_a if result['outcome'] == 'A' else team_b
    loser  = team_b if result['outcome'] == 'A' else team_a

    print(f"  {team_a:<28} {result['win_prob_a']:5.1f}%  vs  "
          f"{result['win_prob_b']:5.1f}%  {team_b:<28} → 🏆 {winner}")

    return winner

# Collect all 32 qualified teams
# Winners and runners-up
w = {g: group_results[g]['ranked'][0] for g in groups}
ru = {g: group_results[g]['ranked'][1] for g in groups}

# Best 8 thirds (already computed)
t3 = [t['team'] for t in best_thirds]

# ── Round of 32 bracket ───────────────────────────────────────────
# FIFA 2026 bracket structure (simplified based on group pathway)
r32_matchups = [
    (w['A'],  t3[0]),   # 1A vs best 3rd #1
    (w['C'],  ru['D']), # 1C vs 2D
    (w['B'],  t3[1]),   # 1B vs best 3rd #2
    (w['D'],  ru['C']), # 1D vs 2C
    (w['E'],  t3[2]),   # 1E vs best 3rd #3
    (w['G'],  ru['F']), # 1G vs 2F
    (w['F'],  t3[3]),   # 1F vs best 3rd #4
    (w['H'],  ru['G']), # 1H vs 2G
    (w['I'],  t3[4]),   # 1I vs best 3rd #5
    (w['K'],  ru['J']), # 1K vs 2J
    (w['J'],  t3[5]),   # 1J vs best 3rd #6
    (w['L'],  ru['K']), # 1L vs 2K
    (ru['A'], t3[6]),   # 2A vs best 3rd #7
    (ru['E'], ru['H']), # 2E vs 2H
    (ru['B'], t3[7]),   # 2B vs best 3rd #8
    (ru['I'], ru['L']), # 2I vs 2L
]

print("\n" + "="*70)
print("  ROUND OF 32")
print("="*70)
r32_winners = []
for a, b in r32_matchups:
    winner = knockout_match(a, b, "R32")
    r32_winners.append(winner)

# ── Round of 16 ───────────────────────────────────────────────────
print("\n" + "="*70)
print("  ROUND OF 16")
print("="*70)
r16_matchups = [
    (r32_winners[0],  r32_winners[1]),
    (r32_winners[2],  r32_winners[3]),
    (r32_winners[4],  r32_winners[5]),
    (r32_winners[6],  r32_winners[7]),
    (r32_winners[8],  r32_winners[9]),
    (r32_winners[10], r32_winners[11]),
    (r32_winners[12], r32_winners[13]),
    (r32_winners[14], r32_winners[15]),
]
r16_winners = []
for a, b in r16_matchups:
    winner = knockout_match(a, b, "R16")
    r16_winners.append(winner)

# ── Quarter Finals ────────────────────────────────────────────────
print("\n" + "="*70)
print("  QUARTER FINALS")
print("="*70)
qf_matchups = [
    (r16_winners[0], r16_winners[1]),
    (r16_winners[2], r16_winners[3]),
    (r16_winners[4], r16_winners[5]),
    (r16_winners[6], r16_winners[7]),
]
qf_winners = []
for a, b in qf_matchups:
    winner = knockout_match(a, b, "QF")
    qf_winners.append(winner)

# ── Semi Finals ───────────────────────────────────────────────────
print("\n" + "="*70)
print("  SEMI FINALS")
print("="*70)
sf_matchups = [
    (qf_winners[0], qf_winners[1]),
    (qf_winners[2], qf_winners[3]),
]
sf_winners = []
sf_losers  = []
for a, b in sf_matchups:
    result = predict_match(a, b)
    if result['outcome'] == 'D':
        winner = a if result['win_prob_a'] >= result['win_prob_b'] else b
    else:
        winner = a if result['outcome'] == 'A' else b
    loser = b if winner == a else a
    sf_winners.append(winner)
    sf_losers.append(loser)
    print(f"  {a:<28} {result['win_prob_a']:5.1f}%  vs  "
          f"{result['win_prob_b']:5.1f}%  {b:<28} → 🏆 {winner}")

# ── Third Place ───────────────────────────────────────────────────
print("\n" + "="*70)
print("  THIRD PLACE PLAYOFF")
print("="*70)
third_place = knockout_match(sf_losers[0], sf_losers[1], "3rd")

# ── Final ─────────────────────────────────────────────────────────
print("\n" + "="*70)
print("  🏆 WORLD CUP FINAL 🏆")
print("="*70)
final_result = predict_match(sf_winners[0], sf_winners[1])
if final_result['outcome'] == 'D':
    champion = sf_winners[0] if final_result['win_prob_a'] >= final_result['win_prob_b'] else sf_winners[1]
else:
    champion = sf_winners[0] if final_result['outcome'] == 'A' else sf_winners[1]
runner_up = sf_winners[1] if champion == sf_winners[0] else sf_winners[0]

print(f"\n  {sf_winners[0]:<28} {final_result['win_prob_a']:5.1f}%  vs  "
      f"{final_result['win_prob_b']:5.1f}%  {sf_winners[1]}")

print(f"\n{'='*70}")
print(f"  🥇 WORLD CUP CHAMPION:    {champion}")
print(f"  🥈 RUNNER UP:             {runner_up}")
print(f"  🥉 THIRD PLACE:           {third_place}")
print(f"{'='*70}")
```
